```python

from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import random
import math

def get_data():
    data = load_svmlight_file("a9a.txt")
    return data[0], data[1]

def get_datat():
    data = load_svmlight_file("a9a.t")
    return data[0], data[1]

def f(x, w, b):
    f_result = 0 #b
    x = np.ndarray.tolist(x)
    for i in range(len(x[0])):
        f_result += x[0][i] * w[i]   #y=wx+b
    f_result += b
    result = 1 / (1 + np.exp(-f_result))
    return result



x_train,y_train=get_data()
x_test,y_test=get_datat()



feature=123
x_train = x_train.todense()
x_test = x_test.todense()



x_train_len=len(x_train)
x_test_len=len(x_test)

for i in range(x_train_len):
    if y_train[i]==-1:
        y_train[i]=0

for i in range(x_test_len):
    if y_test[i]==-1:
        y_test[i]=0


#-------------------------------------NAG-----------------------------------------#
b = 0
w=np.random.rand(123)
lr=0.3
a=0.7
v=np.random.rand(123)*0


loss = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len

count = 0
train_loss = []
test_loss = []
while True:
    k=random.randint(1,32561)
    
    for i in range(122):
       
        gradient_averange = (f(x_train[k,:],w,b)-y_train[k])*x_train[k,i]
      
        v[i] = a*v[i]-lr*gradient_averange #nag
        w[i] += v[i] * lr #nag
        
    train_new_loss = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len

    test_new_loss  = -sum([y_test[i]*math.log(f(x_test[i, :], w, b))+(1-y_test[i])*math.log(1-f(x_test[i, :], w, b))for i in range(x_test_len)])/x_train_len
 





    if abs(loss - train_new_loss) > 0.000001:
        loss = train_new_loss
    else:
        break
    train_loss.append(loss)
    test_loss.append(test_new_loss)
    count += 1
    if count>30:
        break


#-----------------------------------Adadelta-----------------------------------------#

n=np.random.rand(123)*0
m=np.random.rand(123)*0
b=0
p=0.7
w=np.random.rand(123)
lr=0.1
loss2 = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len

count = 0
train_loss2 = []
test_loss2 = []
while True:
    k=random.randint(1,32561)

    for i in range(122):
        
        gradient_averange = (f(x_train[k,:],w,b)-y_train[k])*x_train[k,i]
       
        n[i]= p*n[i]+(1-p)*((gradient_averange)**2)
        w[i] -= lr/math.sqrt(n[i]+0.0000000001)*gradient_averange
        

    train_new_loss2 = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len

    test_new_loss2  = -sum([y_test[i]*math.log(f(x_test[i, :], w, b))+(1-y_test[i])*math.log(1-f(x_test[i, :], w, b))for i in range(x_test_len)])/x_train_len
  
    
    
    
    
    if abs(loss2 - train_new_loss2) > 0.000001:
        loss2 = train_new_loss2
    else:
        break
    train_loss2.append(loss2)
    test_loss2.append(test_new_loss2)
    count += 1
    if count>30:
        break
   


#------------------------------------------RMSProp----------------------------#
n=np.random.rand(123)*0
m=np.random.rand(123)*0
b=0
w=np.random.rand(123)
lr=0.1
p=0.5
loss3 = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len

count = 0
train_loss3 = []
test_loss3 = []
while True:
    k=random.randint(1,32561)
   
    for i in range(122):
        
        gradient_aver = (f(x_train[k,:],w,b)-y_train[k])*x_train[k,i]
    
        n[i]= p*n[i]+(1-p)*((gradient_aver)**2)
        w[i] -= lr/math.sqrt(n[i]+0.0000000001)*gradient_aver
      
    train_new_loss3 = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len

    test_new_loss3  = -sum([y_test[i]*math.log(f(x_test[i, :], w, b))+(1-y_test[i])*math.log(1-f(x_test[i, :], w, b))for i in range(x_test_len)])/x_train_len
    
    
    
    
    
    
    if abs(loss3 - train_new_loss3) > 0.000001:
        loss3 = train_new_loss3
    else:
        break
    train_loss3.append(loss3)
    test_loss3.append(test_new_loss3)
    count += 1
    if count>30:
        break
   

#------------------------------------------Adam----------------------------------#

n=np.random.rand(123)*0
m=np.random.rand(123)*0
m_t=m=np.random.rand(123)*0
n_t=m=np.random.rand(123)*0

p1=0.9
p2=0.99
b = 0
w=np.random.rand(123)
loss4 = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len
lr = 0.0051
cnt = 0
train_loss4 = []
test_loss4 = []
while True:
    cnt += 1
    k=random.randint(1,32561)
    
    for i in range(122):
        
        gradient_aver = (f(x_train[k,:],w,b)-y_train[k])*x_train[k,i]
       
        m[i] = p1*m[i]+(1-p1)*gradient_aver
        #m[i]=lr*math.sqrt(1-p2)/(1-p1)
        n[i] = p2*n[i]+(1-p2)*(gradient_aver**2)
        m_t[i] = m[i]/(1-p1**cnt)
        n_t[i]= n[i]/(1-p2**cnt)
       
        w[i]-= m_t[i]*lr/(math.sqrt(n[i])+0.0000001)
    

    train_new_loss4 = -sum([y_train[i]*math.log(f(x_train[i, :], w, b))+(1-y_train[i])*math.log(1-f(x_train[i, :], w, b))for i in range(x_train_len)])/x_train_len

    test_new_loss4  = -sum([y_test[i]*math.log(f(x_test[i, :], w, b))+(1-y_test[i])*math.log(1-f(x_test[i, :], w, b))for i in range(x_test_len)])/x_train_len

    
    
    
    
    
    if abs(loss4 - train_new_loss4) > 0.000001:
        loss4 = train_new_loss4
    else:
        break

    train_loss4.append(loss4)
    test_loss4.append(test_new_loss4)

    if cnt>30:
        break







plt.figure(figsize=(8,6))
plt.xlabel('Gradient descent - Iteration times')
plt.ylabel('Loss')
plt.plot(range(count), test_loss, 'r-', label=u"testing_nag set")
plt.plot(range(count), test_loss2, 'g-', label=u"testing_adamdelta set")
plt.plot(range(count), test_loss3, 'b-', label=u"testing_rmsprop set")
plt.plot(range(cnt), test_loss4, 'k-', label=u"testing_adam set")


plt.legend()
plt.grid()
plt.show()




















```
